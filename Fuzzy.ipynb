```python
!pip install -U scikit-fuzzy
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

# 1- Membership fn

```python

#-----------------A-Eye-------------------------------
Eye=ctrl.Antecedent(np.arange(0,100,1), 'Temp [EYE %]')
Eye['Pressed closed']=fuzz.trimf(Eye.universe,[0,5,10])
Eye['closed']=fuzz.trimf(Eye.universe,[5,18,30])
Eye['Normal']=fuzz.trimf(Eye.universe,[20,45,70])
Eye['Extra Open']=fuzz.trimf(Eye.universe,[50,75,100])
#-----------------B-Nose-------------------------------
Nose=ctrl.Antecedent(np.arange(0,100,1), 'Temp [Nose %]')
Nose['Wrinked']=fuzz.trimf(Nose.universe,[0,35,60])
Nose['Normal']=fuzz.trimf(Nose.universe,[40,60,100])
#-----------------C-Teath-------------------------------
Teath=ctrl.Antecedent(np.arange(0,100,1), 'Temp [teath %]')
Teath['Not-visible']=fuzz.trimf(Teath.universe,[0,15,30])
Teath['Sightly out']=fuzz.trimf(Teath.universe,[10,40,70])
Teath['Extra Open']=fuzz.trimf(Teath.universe,[50,75,100])
#-----------------D-Lips-------------------------------
Lips=ctrl.Antecedent(np.arange(0,100,1), 'Temp [Lips %]')
Lips['Pressed closed']=fuzz.trimf(Lips.universe,[0,15,30])
Lips['Normal']=fuzz.trimf(Lips.universe,[10,40,70])
Lips['Open']=fuzz.trimf(Lips.universe,[50,75,100])
#-----------------E-Forehead-------------------------------
Forehead=ctrl.Antecedent(np.arange(0,30,1), 'Temp [Forehead %]')
Forehead['Pressed closed']=fuzz.trimf(Forehead.universe,[0,2,4])
Forehead['Normal']=fuzz.trimf(Forehead.universe,[2,8,15])
Forehead['Open']=fuzz.trimf(Forehead.universe,[10,18,30])
#-----------------F-Eyebrows-------------------------------
Eyebrows=ctrl.Antecedent(np.arange(0,100,1), 'Temp [Eyebrows %]')
Eyebrows['Centered']=fuzz.trimf(Eyebrows.universe,[0,10,20])
Eyebrows['Normal']=fuzz.trimf(Eyebrows.universe,[10,35,60])
Eyebrows['Outward-stretched ']=fuzz.trimf(Eyebrows.universe,[40,70,100])
#-----------------G-Cheeks-------------------------------
Cheeks=ctrl.Antecedent(np.arange(0,100,1), 'Temp [Cheeks %]')
Cheeks['Flat&Stretched']=fuzz.trimf(Cheeks.universe,[0,10,20])
Cheeks['Normal']=fuzz.trimf(Cheeks.universe,[10,35,60])
Cheeks['Filled & Up']=fuzz.trimf(Cheeks.universe,[40,70,100])
#-----------------H-Chin-------------------------------
Chin=ctrl.Antecedent(np.arange(0,100,1), 'Temp [Chin %]')
Chin['Normal']=fuzz.trimf(Chin.universe,[0,40,60])
Chin['Redical']=fuzz.trimf(Chin.universe,[40,60,100])

Eye.view()
Nose.view()
Teath.view()
Lips.view()
Forehead.view()
Eyebrows.view()
Cheeks.view()
Chin.view()
```

```python
#----------------------I-Face [outpout]-----------------------
#Face=ctrl.Antecedent(np.arange(0,100,1), 'Temp [output %]')
Face=ctrl.Consequent(np.arange(0,100,1), 'Temp [output %]')
Face ['Anger']=fuzz.trimf(Face.universe,[0,5,10])
Face ['Disgust']=fuzz.trimf(Face.universe,[5,15,25])
Face ['Sad']=fuzz.trimf(Face.universe,[15,28,35])
Face ['Normal']=fuzz.trimf(Face.universe,[30,40,50])
Face ['Happy']=fuzz.trimf(Face.universe,[45,50,60])
Face ['Surprise']=fuzz.trimf(Face.universe,[55,68,75])
Face ['Fear']=fuzz.trimf(Face.universe,[70,77,100])
Face.view()
```

# 2- RULES

```python
#---------------------RULE NUMBER 1--------------------
rul_1=ctrl.Rule(Eyebrows['Centered'] & Lips['Pressed closed'] & Cheeks['Flat&Stretched']&Teath['Not-visible'],Face['Anger'])
#---------------------RULE NUMBER 2--------------------
rul_2=ctrl.Rule(Cheeks['Filled & Up'] &Teath['Sightly out'] &Lips['Open'],Face['Happy'])
#---------------------RULE NUMBER 3--------------------
rul_3=ctrl.Rule(Eyebrows['Normal'] & Lips['Normal'] & Cheeks['Flat&Stretched']&Eye['Normal'],Face['Normal'] )
#---------------------RULE NUMBER 4--------------------
rul_4=ctrl.Rule(Eyebrows['Centered'] &Lips['Normal'] &Cheeks['Flat&Stretched'] ,Face['Sad'])
#---------------------RULE NUMBER 5--------------------
rul_5=ctrl.Rule(Eye['Extra Open']& Lips['Open']&Eyebrows['Outward-stretched '],Face['Surprise'])
#---------------------RULE NUMBER 6--------------------
rul_6=ctrl.Rule(Eyebrows['Centered'] &Nose['Wrinked'] |Cheeks['Filled & Up'],Face['Disgust'])
#---------------------RULE NUMBER 6--------------------
rul_7=ctrl.Rule(Eyebrows['Centered'] &Cheeks['Flat&Stretched']&Eye['closed'] | Eye['Extra Open'] ,Face['Fear'])
```

```python
rul_1.view()
rul_2.view()
rul_3.view()
rul_4.view()
rul_5.view()
rul_6.view()
rul_7.view()
```

```python
tripping_ctrl=ctrl.ControlSystem([rul_1,rul_2,rul_3,rul_4,rul_5,rul_6,rul_7])
tripping=ctrl.ControlSystemSimulation(tripping_ctrl)
```

#3-input DATA

```python
tripping.input['Temp [EYE %]']=7
tripping.input['Temp [Nose %]']=45
tripping.input['Temp [teath %]']= 20
tripping.input['Temp [Lips %]']=15
tripping.input['Temp [Eyebrows %]']=15
tripping.input['Temp [Cheeks %]']=10
```

4- COMPUTE

```python
tripping.compute()
```

5-OUTPUT

```python
print(tripping.output['Temp [output %]'])
Face.view(sim=tripping)
```
